```python
from equityiq.report import Report

report = Report("AAPL")
print(report.summary())
```

```python
tickers = ["AAPL", "MSFT", "GOOGL"]
dataframes = []

for t in tickers:
    r = Report(t)
    df = r.to_dataframe()
    df["ticker"] = t
    dataframes.append(df)

comparison = pd.concat(dataframes).pivot_table(
    index="metric", columns="ticker", values="value", aggfunc="first"
)
print(comparison)
```

```python
from equityiq.projection import Projector

# Escenario 1: usando medias históricas como default
p = Projector("META")
p.summary()

# Escenario 2: introduciendo tus propios supuestos (como en el Excel)
p2 = Projector("META", revenue_growth=0.20, ebit_margin=0.40, tax_rate=0.15)
p2.summary()
```

```python
from equityiq.projection import Projector
from equityiq.multiples_valuation import MultiplesValuation
# Un solo valor para todos los años
p = Projector("META", revenue_growth=0.20, ebit_margin=0.40, tax_rate=0.15)
mv = MultiplesValuation("META", projector=p, ev_fcf_target=25, per_target=25,
                         ev_ebitda_target=17, ev_ebit_target=20)
mv.summary()
```

```python
# Año 1 diferente, resto igual
p = Projector("META")
mv = MultiplesValuation("META", projector=p, ev_fcf_target=25, per_target=25,
                         ev_ebitda_target=17, ev_ebit_target=20)
mv.summary()
```

```python
# Valor distinto para cada año
p = Projector("META",
              revenue_growth=[0.258, 0.188, 0.168, 0.143, 0.139],
              ebit_margin=[0.354, 0.346, 0.346, 0.369, 0.373],
              tax_rate=0.15)
mv = MultiplesValuation("META", projector=p, ev_fcf_target=25, per_target=25,
                         ev_ebitda_target=17, ev_ebit_target=20)
mv.summary()
```
